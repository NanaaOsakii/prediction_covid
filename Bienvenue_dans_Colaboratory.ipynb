```
import pandas as pd

data = "/content/Défaillance client.xlsx"
feuilles = ["Encaissement annuel 2022", "Encaissement annuel 2023", "CA annuel 2022", "CA annuel 2023"]

xl = pd.ExcelFile(data)

for nom_feuille in feuilles:
    if nom_feuille in xl.sheet_names:
        df = xl.parse(nom_feuille)
        print("\nContenu de la feuille :", nom_feuille)
        print(df)
    else:
        print("\nLa feuille", nom_feuille, "n'existe pas dans le fichier Excel.")
```
